```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(rc={'figure.figsize':[8,8]},font_scale=1.4)
```

```python
df=pd.read_csv('911.csv')
df.head(6)
```

```python
df.info()
```

```python
sns.countplot(x='hour', data=df, palette='viridis')
```

```python
sns.countplot(x='month', data=df, palette='mako')
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,5),sharey=True)
fig.suptitle('trouble density')
sns.countplot(x='Year',ax=axes[0],data=df)
sns.countplot(x='Reason',ax=axes[1],data=df)
```

```python
sns.countplot(x='Year',data=df,hue='Reason',palette='mako')
```

```python
sns.countplot(x='Year',data=df,hue='day of week')
```
