## Titanic Analysis

###### survived, sex, age, fare, class, who, embark_town, alive, alone

```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.set(rc={'figure.figsize':[10,10]},font_scale=1.5)
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
sns.distplot(df['age'],color='m')
```

```python
sns.kdeplot(df['fare'],shade=True,color='m')
```

```python
sns.jointplot(x='age', y='fare', data=df, color='r')
```

```python
sns.countplot(x='alive',data=df,palette='mako')
```

```python
sns.countplot(x='class',data=df,hue='who',palette='mako')
```

```python
sns.countplot(x='who',data=df,hue='alive')
```

```python
sns.countplot(x='class', data=df, hue='who', palette='flare')
```

```python
sns.boxplot(x='survived', y='age', data=df)
```

```python
sns.boxplot(x='sex',y='age',data=df,hue='alone',palette='mako')
```

```python
sns.violinplot(x='class', y='age', data=df)
```

```python
sns.violinplot(x='class', y='age', data=df, hue='survived', split=True)
```

```python
sns.stripplot(x='sex', y='age', data=df, hue='survived', dodge=True)
```

```python
sns.swarmplot(x='sex', y='age', data=df, hue='survived', dodge=True)
```

```python
sns.swarmplot(x='embark_town', y='age', data=df, hue='survived', dodge=True)
```

```python
sns.barplot(x='class', y='fare', data=df, estimator=np.sum, hue='embark_town')
```

```python
df_corr = df.corr()
df_corr
```

```python
sns.heatmap(df_corr, cmap='viridis', linecolor='k', linewidths=2 ,annot=True,)
```
